Iteration: Simple Population Growth Model

n = population 
<br>
n_0 = intital popultation
<br>
n_1 = population at year 1
<br>
n_t = pop at year t
<br>
birthrate (br) =  number of offspring produced each year
<br>
n_1 = br * n_0
<br>
n_2 = br * n_1
<br>
k = Max_Population (carrying capacity)
<br>
dr = deathrate

General model<br>

n_t + 1 = br * n_t

<br>
Extended (Logistic) Model:
<br>
n_t+1 = (br - dr)[n_t - nt**2 / k]



```python
#initial settings
int_pop = 1.0
birth_rate = 2.0
death_rate = 0.0
carrying_capacity = 50.0
time_step = 10.0

new_pop = (birth_rate - death_rate) * (int_pop - (int_pop**2 / carrying_capacity))
print (new_pop)
```

```python
def Population(int_pop, birth_rate, death_rate, carrying_capacity, time_step):
    
    out = np.array([])
    out = np.append(out, int_pop)
    last_pop = np.array([])
    last_pop = np.append(last_pop, int_pop)  # track last years pop
    while time_step > 0:
        new_pop = (birth_rate - death_rate) * (int_pop - (int_pop**2 / carrying_capacity))
        print (new_pop)
        out = np.append(out, [new_pop])
        last_pop = np.append(last_pop, [int_pop])
        int_pop = new_pop
        time_step = time_step - 1
    
    return out, last_pop
```

```python
#output
bunnie_count, last_years_bunnies = (Population(int_pop, birth_rate, death_rate, carrying_capacity, time_step))
print (bunnie_count) 
print (last_years_bunnies)
```

This year population is Bunnie Count

```python
plt.plot(bunnie_count)
plt.title('Simple Population growth')
plt.ylabel('population')
plt.xlabel('Time')
plt.ylim(ymin=0)  
plt.xlim(xmin=0)
plt.show
```

```python
plt.title('Last Year Growth v This years')
plt.ylabel('This Year')
plt.xlabel('Last Year')

plt.plot(last_years_bunnies,bunnie_count)
plt.show()

print ('bunnie_count      ', bunnie_count) 
print ('last_years_bunnies',last_years_bunnies)
```
